```python
Import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame
from ipywidgets import Dropdown, FloatSlider, IntSlider, interact
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
#Installation of Category_Encoders on Google Collab#
! pip install category_encoders
from category_encoders import OneHotEncoder


```

```python
#Import the Data#
df = pd.read_csv('Prediction of PAT 2022.csv')

#Split the Data#
features = ["total_assets", "customer_deposits", "loans_advances", "Bank"]
target = "pat"
X_train = df[features]
y_train = df[target]

#Make a model and rub it through One_Hot_Encoder and Ridge#
model = make_pipeline(
    OneHotEncoder(use_cat_names=True),
    Ridge()
)

#Fit the model unto our data#
model.fit(X_train, y_train)

#Make Our predictions using a function#
def make_prediction(total_assets, customer_deposits, loans_advances, Bank):

    predictions = {
        "total_assets":total_assets,
        "customer_deposits":customer_deposits,
        "loans_advances":loans_advances,
        "Bank":Bank
    }
    df = pd.DataFrame(predictions, index=[0])
    prediction = model.predict(df).round(4)
    return f"Predicted Profit After Tax: shs{prediction}"
```

```python
#Alternative to Make predictions using Intslider#
interact(
    make_prediction,
    total_assets=IntSlider(
        min=X_train['total_assets'].min(),
        max=X_train['total_assets'].max(),
        value=X_train['total_assets'].mean(),
    ),
    customer_deposits=FloatSlider(
        min=X_train['customer_deposits'].min(),
        max=X_train['customer_deposits'].max(),
        step=0.01,
        value=X_train['customer_deposits'].mean(),
    ),
    loans_advances=FloatSlider(
        min=X_train['loans_advances'].min(),
        max=X_train['loans_advances'].max(),
        step=0.01,
        value=X_train["loans_advances"].mean(),
    ),
    Bank=Dropdown(options=sorted(X_train['Bank'].unique())),
);
```
