IMPORT LIBRARIES

```python
import pandas as pd
import random
```

IMPORT CODE


```python
df = pd.read_csv('data.csv')
df.head()
```

BUILD CODE

```python
def random_values(df, column_name):
  return random.choice(df[column_name].tolist())

Randomization = random_values(df, 'STATE')
print('Random state that has been picked', Randomization)
```
