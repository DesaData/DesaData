```python
#Methodology used in analysing the data#
#1)Importing of Data#
#2) Dropping of Unnecessary columns#
#3) Changing the financials to Float#
#4) Adding columns for Percentage change#
#5) Analysis#
#6) Visualization#

#1)Importing of Data#
import pandas as pd
import numpy as np
df = pd.read_csv('MicrosofTeslaApple.csv')
df.info()
```

```python
#2) Dropping of Unnecessary columns#
df.drop(columns=['Unnamed: 8'], inplace=True)
df.drop(columns=['Unnamed: 9'], inplace=True)
df.drop(columns=['Unnamed: 10'], inplace=True)
df.drop(columns=['Unnamed: 11'], inplace=True)
df.drop(columns=['Unnamed: 12'], inplace=True)
df.drop(columns=['Unnamed: 13'], inplace=True)
df.drop(columns=['Unnamed: 14'], inplace=True)
df.drop(columns=['Unnamed: 15'], inplace=True)
df.drop(columns=['Unnamed: 16'], inplace=True)
df.drop(columns=['Unnamed: 17'], inplace=True)
df.drop(columns=['Unnamed: 18'], inplace=True)
df.drop(columns=['Unnamed: 19'], inplace=True)
df.drop(columns=['Unnamed: 20'], inplace=True)
df.drop(columns=['Unnamed: 21'], inplace=True)
df.drop(columns=['Unnamed: 22'], inplace=True)
df.drop(columns=['Unnamed: 23'], inplace=True)
df.drop(columns=['Unnamed: 24'], inplace=True)
df.drop(columns=['Unnamed: 25'], inplace=True)
df.drop(columns=['Unnamed: 26'], inplace=True)
```

```python
df.head()
```

```python
#3) Changing the financials to Float#
df['Revenue'] = df['Revenue'].str.replace(',',' ',regex=False)
df['Revenue'] = df['Revenue'].str.replace('$',' ',regex=False)
df['Revenue'] = df['Revenue'].str.replace(' ','',regex=False)
df['Revenue'] = df['Revenue'].astype(float)
df['Net Income'] = df['Net Income'].str.replace(',',' ',regex=False)
df['Net Income'] = df['Net Income'].str.replace('$',' ',regex=False)
df['Net Income'] = df['Net Income'].str.replace(' ','',regex=False)
df['Net Income'] = df['Net Income'].astype(float)
df['Total Assets'] = df['Total Assets'].str.replace('$',' ',regex=False)
df['Total Assets'] = df['Total Assets'].str.replace(',',' ',regex=False)
df['Total Assets'] = df['Total Assets'].str.replace(' ','',regex=False)
df['Total Assets'] = df['Total Assets'].astype(float)
df['Total Liabilities'] = df['Total Liabilities'].str.replace(',',' ')
df['Total Liabilities'] = df['Total Liabilities'].str.replace('$',' ')
df['Total Liabilities'] = df['Total Liabilities'].str.replace(' ','')
df['Total Liabilities'] = df['Total Liabilities'].astype(float)
df['Cash flow'] = df['Cash flow'].str.replace(',',' ')
df['Cash flow'] = df['Cash flow'].str.replace('$',' ')
df['Cash flow'] = df['Cash flow'].str.replace(' ','')
df['Cash flow']= df['Cash flow'].str.replace('\$,',' ')
df.head()
```

```python
df.info()
```

```python
# 4) Adding columns for Percentage change#
df['Revenue Growth (%)'] = df.groupby('Company')['Revenue'].pct_change() * 100
df['Net Income (%)'] = df.groupby('Company')['Net Income'].pct_change() * 100
df['Total Assets (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Total Liabilities (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df.head(9)
```

```python
#5) Analysis#
#Revenue Growth# 1) Microsoft had an upward growth of revenue from 2021 to 2023 with an overall 6.8 increase#
                #2) Tesla had a slump in revenue dropping by 18% from 2022#
                #3)Apple  faced a major decline in revenue dropping to -7% in 2023#
#Net Income# 1) Microsoft faced a staedy decline in net income from 2021 to 2023 to -0.5%#
            #2)Tesla's net income declined more than half over the period from#
#Total Assets #1) Microsoft's total Assets increased from 2021 to 2023  by 12.1% increase from 2021 to 2022#
              #2) Tesla's assets increased with the biggest notable change between 2021 and 2022#
              #3)Apple's  assets barely had any significant change and remained within the same period#
#Total Liabilities 1) Microsoft liabilities slightly increased especially from 2022 to 2023 citing a 3.7% increase#
                  #2)Tesla's liabilities gradually increased over the 3 year period from 2021 to 2023#
                  #3)Apple 's liabilities reduced gradually with a 4% decline from 2021 to 2023
```

```python
#6) Visualization#
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Net Income', data=df, label='Net Income', color='blue')
sns.lineplot(x='Year', y='Revenue', data=df, label='Revenue', color='green')
sns.lineplot(x='Year', y='Total Liabilities', data=df, label='Total Liabilities',color='red')
sns.lineplot(x='Year', y='Total Assets', data=df, label='Total Assets',color='blue')
plt.xlabel('Year')
plt.ylabel('Financials')
plt.show()
```
